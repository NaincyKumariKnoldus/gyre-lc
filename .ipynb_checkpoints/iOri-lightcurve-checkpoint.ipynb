```python
class star:
    
    def __init__ (self, inlist_path, x, use_grid=True, star_number=1):
        
        # should check if 'inlist', this just checks 'str'
        
        if isinstance(inlist_path, str):
            star_inlist = nml.read(inlist_path)[f'star_{star_number}'] #Replace 'star number' with an external specification of the star, so that later I can use it for single stars
            observer_inlist = nml.read(inlist_path)[f'observer']
        else: raise Exception('Inlist file error')
            
        self.params = self.read_mesa_params(star_inlist['star_model_path'])
        
        self.x = x
        self.use_grid = use_grid
        
        self.resp_coeffs = rc.resp_coeffs(star_inlist['tide_model_path'])
        
        if use_grid==False:
            self.phot_coeffs = self.read_atm_coeffs('lc-data/iOri-partials.h5') 
        else:
            self.phot_coeffs = self.read_pg_coeffs()
        
            
    def read_mesa_params(self, star_model_path, units='SOLAR'):
        
        data = ascii.read(star_model_path, data_start=0, data_end=1)
        
        # cgs constants
        
        R_sol = 6.957e10 
        M_sol = 1.989e33
        L_sol = 3.839e33
        G = 6.674079999999999e-08 
        sigma_sb = 5.6703669999999995e-05
        
        # data in cgs
        
        m = data[0][1]
        r = data[0][2]
        l = data[0][3]
            
        # calculate Teff [K], logg [dex]
        
        Teff = (l/(sigma_sb * 4*np.pi*r**2))**0.25
        
        g_surf = G*m/(r**2)
        
        logg = np.log10(g_surf)
        
        # convert to solar units
        
        if units=='SOLAR':
            m = m/M_sol
            r = r/R_sol
            l = l/L_sol
            
        return {'M': m,
                'R': r,
                'L': l,
                'Teff': Teff,
                'logg': logg}
    
    
    def read_atm_coeffs(self, phot_file):
        # This function is old and uses a synspec
        # intensity spectrum calculated from iOri's 
        # tlusty atmosphere.
        
        l_min = 0
        dl = 1
        l_max = self.resp_coeffs.data['l_max']
        n_l = np.ceil( (l_max - l_min)/dl ) + 1

        l_range = l_min + dl*np.arange(n_l)
        
        # read intensity file
        
        I = ac.atm_coeffs(phot_file)
        I = I.data
        
        # Evaluate photometric data
        I_x_l = {}
        dI_dlnT_x_l = {}
        dI_dlng_x_l = {}
        
        for l in l_range.astype(int):
            I_x_l[l] = I[f'I_{self.x}_{l}'][0]
            dI_dlnT_x_l[l] = I[f'dlnTeff_{self.x}_{l}'][0]
            dI_dlng_x_l[l] = I[f'dlng_{self.x}_{l}'][0]
            
        return {f'I_{self.x}': I_x_l, f'dI_dlnT_{self.x}': dI_dlnT_x_l, f'dI_dlng_{self.x}': dI_dlng_x_l}
        
            
    def read_pg_coeffs(self): 
        #'x' should be specified in the inlist
        # and *only* upon loading the msg PhotGrid
        
        # Set atmosphere parameters dict
    
        Teff = self.params['Teff']
        logg = self.params['logg']

        dx = {'logT': np.log10(Teff), 'logg': logg}
        pg = pymsg.PhotGrid(f'pg-demo-{self.x}.h5') #f'pg-demo-ostar2002.h5'
        
        print(dx, f'Teff: {Teff}')
        
        # Set up intensity moment range
        
        l_min = 0
        dl = 1
        l_max = self.resp_coeffs.data['l_max']
        n_l = np.ceil( (l_max - l_min)/dl ) + 1

        l_range = l_min + dl*np.arange(n_l)
        
        # Evaluate photometric data
        I_x_l = {}
        dI_dlnT_x_l = {}
        dI_dlng_x_l = {}
        
        for l in l_range.astype(int):
            I_x_l[l] = pg.D_moment(dx, l)
            dI_dlnT_x_l[l] = pg.D_moment(dx, l, deriv={'logT':True})/np.log(10.)
            dI_dlng_x_l[l] = pg.D_moment(dx, l, deriv={'logg':True})/np.log(10.)
            
        return {f'I_{self.x}': I_x_l, f'dI_dlnT_{self.x}': dI_dlnT_x_l, f'dI_dlng_{self.x}': dI_dlng_x_l}
    
    
    def R_xl(self, l):
        if self.use_grid==True:
            Teff = self.params['Teff']
            logg = self.params['logg']

            dx = {'logT': np.log10(Teff), 'logg': logg}
            pg = pymsg.PhotGrid(f'pg-demo-{self.x}.h5')
            
            I = pg.D_moment(dx, l)
            return (2 + l)*(1 - l)*I / pg.D_moment(dx, 0)
        else:
            I = self.phot_coeffs
            return (2 + l)*(1 - l)*I[f'I_{self.x}'][l] / I[f'I_{self.x}'][0]
    '''
    def T_xl(self, x, l):
        if x=='kepler':
            I = pg.D_moment(dx, l)
        else:
            I = self.phot_coeffs
            return I[f'dI_dlnT_{x}'][l] / I[f'I_{x}'][0] 
    
    def G_xl(self, x, l):
        
        if x=='kepler':
            I = pg.D_moment(dx, l)
        else:
            I = self.phot_coeffs
            return I[f'dI_dlng_{x}'][l] / I[f'I_{x}'][0]
                    
    '''
    #def atm_coeffs(self, x, l, m, k):
    #    return
```

```python
class binary:
    
    def __init__ (self, inlist_path, x, use_grid=True):
        
        # should check if 'inlist', this just checks 'str'
        
        if isinstance(inlist_path, str):
            self.inlist = nml.read(inlist_path)
        else: raise Exception('Inlist file error')
        
        self.component = {1: star(inlist_path, x, use_grid, 1),\
                          2: star(inlist_path, x, use_grid, 2) }
        
        self.orbit_params = self.read_orbit()
            
        
    def read_orbit(self):
        
        orbit_inlist = self.inlist['orbit']
        
        params = {}
        
        #placeholder unit routine
        
        if orbit_inlist['omega_orb_units']=='CYC_PER_DAY':
            omega_orb_units = 1 
        else: omega_orb_units = 1
        
        if orbit_inlist['a_units']=='CM':
            a_units = 6.957e10 #cm per r_sol
        else: a_units = 1
            
        # eventually, we want to get these from resp_data
        # resp_data currently does not contain this output tho
        
        params['a'] = orbit_inlist['a']*a_units
        params['e'] = orbit_inlist['e']
        params['Omega_orb'] = orbit_inlist['omega_orb']*omega_orb_units
        
        return params
   
```

```python
# under construction
class component(star):
    
    def __init__ (self, inlist_path, x, use_grid=True, star_number=1):
        super().__init__(inlist_path, x, use_grid, star_number)
        # prep for find_bin_sep()
        
        #self.e = binary_object.orbit['e']
        #self.a = binary_object.orbit['a']
        #self.Omega_orb = binary_object.orbit['Omega_orb']
        
        # prep for eval_irrad()
        
        #if int(star_number)==1:
        #    star_neighbor = 2
        #elif int(star_number)==2:
        #    star_neighbor = 1
        #else:
        #    raise Exception('Star unspecified')
        
        #self.atm_data = bin_data.star[star_number].atm
        #self.resp_data = {1: binary_object.star[1].resp_coeffs.data,
        #                  2: binary_object.star[2].resp_coeffs.data}
        
        #self.L = {1: binary_object.star[1].params['L'],
        #          2: binary_object.star[2].params['L']}
        #self.R = {1: binary_object.star[1].params['R'],
        #          2: binary_object.star[2].params['R']}
        
        
        self.ramp = self.eval_ramp()
        self.disk_intg_factor = self.eval_disk_intg_factor( x )
        
        # star 1
       
    
    def find_ramp (self, l, m):
        return(self.ramp[l][m])
    
            
    def eval_ramp (self):
        
        # find photometric moment range
        
        l_min = 0
        dl = 1
        l_max = resp_coeffs.data['l_max']
        n_l = np.ceil( (l_max - l_min)/dl ) + 1

        l_range = l_min + dl*np.arange(n_l)
  
        # initialize dictionary
    
        Z_ = {}
  
        # iterating through the elements of list
    
        for l in l_range.astype(int):
            Z_[l] = {}
            
            for m in np.arange(-l, l+1).astype(int):
                # Could this be calling m, l wrong?
                
                term1 = 2* np.sqrt( ((2*l+1)/(4*np.pi)) \
                           *(np.math.factorial(l-m)/np.math.factorial(l+m)) )
                
                if (np.abs(m)==1):
                    term2 = np.pi/2
                else:
                    term2 = np.cos(m*np.pi/2)/(1-m**2)
        
                term3, term3_err = integrate(lambda mu: np.sqrt(1-mu**2)*lpmv(m,l, mu), -1, 1)
        
                Z_[l][m] = term1*term2*term3
        
        return Z_
    
    
    def find_disk_intg_factor (self, l):
        return(self.disk_intg_factor[l])
    
    
    def eval_disk_intg_factor (self, x):
        
        l_min = 0
        dl = 1
        l_max = self.resp_coeffs.data['l_max']
        n_l = np.ceil( (l_max - l_min)/dl ) + 1

        l_range = l_min + dl*np.arange(n_l)
        
        # initialize dictionary
    
        b_ = {}
  
        # iterating through the elements of list
    
        for l in l_range.astype(int):
            b_[l] = self.phot_coeffs[f'I_{x}'][l]/self.phot_coeffs[f'I_{x}'][0]
    
        return b_
        
    
    def find_mean_anom (self, binary_object, t, t_peri=0):
        
        return binary_object.params['Omega_orb']*(t - t_peri)*(2*np.pi)
        
    
    def find_ecce_anom (self, e, M):
    
        K_soln = np.empty_like(M)
        
        for i, M_val in enumerate(M):
            
            Keppler = lambda E : E - e*np.sin(E) - M_val
            K_soln[i] = fsolve(Keppler, 0)[0]
        
        return K_soln
        
    
    def find_true_anom (self, E):
    
        return 2*np.arctan( ((1+self.e)/(1-self.e))*np.tan(E/2) )
    
    
    def convert_t_to_f (self, t, t_peri=0):
    
        M = self.find_mean_anom(t, t_peri)
        
        E = self.find_ecce_anom(M)
        
        f = self.find_true_anom(E)
        
        return f
    
    
    def find_bin_sep (self, t, t_peri=0):
        
        f = self.convert_t_to_f(t, t_peri)#* 2*np.pi
        
        D = self.a*(1-self.e**2)/(1+self.e*np.cos(f))
        
        return D
    
    
    def find_irrad (self, x, t, t_peri=0):
        
        Dt = self.find_bin_sep(t, t_peri)
        
        #rel_dJm = np.zeros(self.res_data['l_max']+1, dtype=complex)
        
        rel_dJ = np.zeros_like(t)
        
        # Set up intensity moment range
        
        l_min = 0
        dl = 1
        l_max = self.res_data['l_max']
        n_l = np.ceil( (l_max - l_min)/dl ) + 1

        l_range = l_min + dl*np.arange(n_l)
        
        for l in l_range.astype(int):
            for m in np.arange(-l, l+1).astype(int):
                
                Z_lm = self.find_ramp(l, m)
                b_l = self.find_disk_intg_factor(l)
                rel_dJ += b_l*Z_lm*(self.L2/self.L1)*(self.R1/Dt)**2
        
        return rel_dJ
        
            #dJ/unperturbed_observed_flux 
            # = bandpass_correction_coefficient*(L2/L1)*(R1/D(t))**2
            #   * disc_integral_factor_(Burkart_used_Eddington_limb_darkening)
            
```

```python
class irradiation:
    
    def __init__ (self, bin_data, star_number):
        
        # prep for find_bin_sep()
        
        self.e = bin_data.orbit_params['e']
        self.a = bin_data.orbit_params['a']
        self.Omega_orb = bin_data.orbit_params['Omega_orb']
        
        # prep for eval_irrad()
        
        if int(star_number)==1:
            star_neighbor = 2
        elif int(star_number)==2:
            star_neighbor = 1
        else:
            raise Exception('Star unspecified')
        
        self.atm_data = bin_data.component[star_number].phot_coeffs
        self.res_data = bin_data.component[star_number].resp_coeffs.data
        self.x = bin_data.component[star_number].x
        
        self.L1 = bin_data.component[star_number].params['L']
        self.R1 = bin_data.component[star_number].params['R']
        self.L2 = bin_data.component[star_neighbor].params['L']
        
            
    def eval_ramp (self, l, m):
        
        term1 = 2* np.sqrt( ((2*l+1)/(4*np.pi)) \
                           *(np.math.factorial(l-m)/np.math.factorial(l+m)) )
        
        if (np.abs(m)==1):
            term2 = np.pi/2
        else:
            term2 = np.cos(m*np.pi/2)/(1-m**2)
        
        term3, term3_err = integrate(lambda mu: np.sqrt(1-mu**2)*lpmv(m,l, mu), -1, 1)
        
        Z_lm = term1*term2*term3
        
        return Z_lm
    
    
    def find_disk_intg_factor (self, l):
    
        b_l = self.phot_coeffs[f'I_{self.x}'][l]/self.phot_coeffs[f'I_{self.x}'][0]
    
        return b_l
        
    
    def find_mean_anom (self, t, t_peri=0):
        
        return self.Omega_orb*(t - t_peri)*(2*np.pi)
        
    
    def find_ecce_anom (self, M):
    
        K_soln = np.empty_like(M)
        
        for i, M_val in enumerate(M):
            
            Keppler = lambda E : E - self.e*np.sin(E) - M_val
            K_soln[i] = fsolve(Keppler, 0)[0]
        
        return K_soln
        
    
    def find_true_anom (self, E):
    
        return 2*np.arctan( ((1+self.e)/(1-self.e))*np.tan(E/2) )
    
    
    def convert_t_to_f (self, t, t_peri=0):
    
        M = self.find_mean_anom(t, t_peri)
        
        E = self.find_ecce_anom(M)
        
        f = self.find_true_anom(E)
        
        return f
    
    
    def find_bin_sep (self, t, t_peri=0):
        
        f = self.convert_t_to_f(t, t_peri)#* 2*np.pi
        
        D = self.a*(1-self.e**2)/(1+self.e*np.cos(f))
        
        return D
    
    
    def find_irrad (self, t, t_peri=0):
        
        Dt = self.find_bin_sep(t, t_peri)
        
        rel_dJ = np.zeros_like(t)
        
        # Set up intensity moment range
        
        l_min = 0
        dl = 1
        l_max = self.res_data['l_max']
        n_l = np.ceil( (l_max - l_min)/dl ) + 1

        l_range = l_min + dl*np.arange(n_l)
        
        for l in l_range.astype(int):
            for m in np.arange(-l, l+1).astype(int):
                
                Z_lm = self.eval_ramp(l, m)
                b_l = self.find_disk_intg_factor(l)
                rel_dJ += b_l*Z_lm*(self.L2/self.L1)*(self.R1/Dt)**2
        
        return rel_dJ
        
            #dJ/unperturbed_observed_flux 
            # = bandpass_correction_coefficient*(L2/L1)*(R1/D(t))**2
            #   * disc_integral_factor_(Burkart_used_Eddington_limb_darkening)
            
```

```python
class observer:
    
    def __init__ (self, list_path, x, use_grid=True):
        
        if isinstance(list_path, str):
            self.nml = nml.read(list_path)
        else: raise Exception('Inlist file error')
        
        self.use_grid = use_grid
        self.bin_data = binary(list_path, x, use_grid)
        self.x = x
        
        # Note: if run from terminal, all timeseries functions are run at execution. 
        # If run from jupyter, they can accept any user-specified times & inclination (but not filter yet)
        self.irr = {1: irradiation(self.bin_data, 1),\
                    2: irradiation(self.bin_data, 2)}
        
    
    def convert_coords (self, inc, omega):

        theta = inc/180 * np.pi
        phi = (90-omega)/180 * np.pi
        
        return theta, phi
        
    
    def calc_bol_I (self, I_0, l):
        
        if l == 0:
            I_l = 1
        elif l == 1:
            I_l = 1/3
        elif l == 2:
            I_l = 1/8
        elif l == 3:
            I_l = 0
        elif l == 4:
            I_l = -1/48
        elif l == 5:
            I_l = 0
        elif l == 6:
            I_l = 1/128
        else:
            raise Exception('No I_l defined')
            
        return I_l*I_0
            
    
    def eval_fourier (self, star_number, inc, omega, bol=False):
        
        res = self.bin_data.component[star_number].resp_coeffs
        atm = self.bin_data.component[star_number].phot_coeffs
        
        theta, phi = self.convert_coords(inc, omega)
        
        # Initialize the frequencies/amplitudes arrays
        
        f = np.arange(res.data['k_max']+1)*res.data['Omega_orb']
        
        A = np.zeros(res.data['k_max']+1, dtype=complex)
            

        # Loop over l, m and k
        I = atm
    
        for l in np.arange(2, res.data['l_max']+1).astype(int):
            for m in np.arange(-l, l+1).astype(int):
                for k in np.arange(0, res.data['k_max']+1).astype(int):

                    i_l = l
                    i_m = m + res.data['l_max']
                    i_k = k

                    dR_lmk = res.R_lmk[i_l,i_m,i_k]
                    dT_lmk = res.T_lmk[i_l,i_m,i_k]
                    dG_lmk = res.G_lmk[i_l,i_m,i_k]

                    Y_lm = sph_harm(m, l, phi, theta)
                    
                    R_xl = (2 + l)*(1 - l)*I[f'I_{self.x}'][l] / I[f'I_{self.x}'][0]
                    T_xl = I[f'dI_dlnT_{self.x}'][l] / I[f'I_{self.x}'][0] 
                    G_xl = I[f'dI_dlng_{self.x}'][l] / I[f'I_{self.x}'][0]
                    
                    if bol==True:
                        I_l = self.calc_bol_I( I[f'I_{self.x}'][0], l)
                        R_xl = (2 + l)*(1 - l)*I_l / I[f'I_{self.x}'][0]
                        T_xl = 4*I_l / I[f'I_{self.x}'][0]
                        G_xl = 0
                    
                    if k == 0:
                        if m == 0:
                            kappa = 0.5
                        elif m >= 1:
                            kappa = 1.
                        else:
                            kappa = 0.
                    else:
                        kappa = 1.
    
                    # Add the Fourier contribution * spherical harmonic
                    
                    A[i_k] += 2*kappa*(dR_lmk*R_xl)*Y_lm # + dT_lmk*T_xl + dG_lmk*G_xl
        
        # Return data
        return f, A
    

    def find_star_flux (self, star_number, inc, omega, t, t_peri=0, bol=False):
        
        res = self.bin_data.component[star_number].resp_coeffs
        Omega_orb = self.bin_data.orbit_params['Omega_orb']
        
        f, A = self.eval_fourier(star_number, inc, omega, bol)
        
        # Initialize the frequencies/amplitudes arrays
        
        if isinstance(t, np.ndarray):
            
            star_flux = np.zeros_like(t)
        
        elif isinstance(t, list):
            
            t = np.array(t)
            star_flux = np.zeros_like(t)
            
        else: #How to isinstance any mathable t? float, int, etc?
            
            star_flux = 0.

        # Add contributions from each frequency component

        n_A = len(A)

        for i in range(n_A):
            
            star_flux += np.real(A[i] * np.exp(-1j*f[i]*2*np.pi*(t - t_peri)))
            
        return star_flux
    
    
    def find_flux (self, inc, omega, t, t_peri=0, reflection=True, bol=False):
        
        star_1 = self.find_star_flux(1, inc, omega, t, t_peri, bol)
        star_2 = self.find_star_flux(2, inc, omega, t, t_peri, bol)
        
        if reflection==False:
            return star_1 + star_2
        
        else:
            irr_1 = self.irr[1].find_irrad(t, t_peri)
            irr_2 = self.irr[2].find_irrad(t, t_peri)
        
            return irr_1 + irr_2 + star_1 + star_2
    
```
